```python
import os
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
gpd.show_versions()
```

```python
coastlines_path = os.path.join("ne_50m_coastline",
                               "ne_50m_coastline.shp")
coastlines = gpd.read_file(coastlines_path)
coastlines.head()
```

```python
type(coastlines),coastlines.shape
```

```python
coastlines['geometry'][55]
```

```python
f,ax1 = plt.subplots(figsize=(12,6))
coastlines.plot(ax=ax1)
ax1.set(title='Global coastlines Boundaries')
plt.savefig('Global coastlines Boundaries')
plt.show()
```
